```python
new_df = df.drop(['American Joint Committee on Cancer Publication Version Type','American Joint Committee on Cancer Metastasis Stage Code','Primary Lymph Node Presentation Assessment','Patient Weight'], axis = 1)
```

```python
new_df.head()
```

```python
final_df = new_df
for col in final_df.columns:
  if(len(final_df[col].unique()) == 1):
    final_df = final_df.drop(col, axis = 1)
```

```python
final_df.head()
```

```python
final_df.to_csv('final_df')
```

```python
final_df.dropna(subset = ['Disease Free (Months)', 'Disease Free Status'], inplace = True)
final_df
```

```python
df = final_df
df.loc[df['Disease Free Status'] == '0:DiseaseFree', 'Disease Free Status'] = 0
df.loc[df['Disease Free Status'] == '1:Recurred/Progressed', 'Disease Free Status'] = 1
df['Disease Free Status'].unique()
```

```python
df.head()
```

```python
df = df.drop(['Months of disease-specific survival', 'Disease-specific Survival status', 'Form completion date'], axis = 1)
```

```python
from numpy import NaN
df.loc[df['Ethnicity Category'] == 'Not Hispanic Or Latino', 'Ethnicity Category'] = 1
df.loc[df['Ethnicity Category'].isna() , 'Ethnicity Category'] = 0
```

```python
df = df.drop(['ICD-10 Classification'], axis = 1)
```
